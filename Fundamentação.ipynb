```python
zip_file_name = "/content/2024.zip"
import pandas as pd
import zipfile

with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    with zip_ref.open('INMET_S_SC_A895_CHAPECO_01-01-2024_A_30-06-2024.CSV') as csv_file:

        df = pd.read_csv(csv_file, sep=';',skiprows=8, encoding='latin-1',on_bad_lines='skip')
```

```python
for column in df:
  if df[column].dtype == 'object':
    df[column] = df[column].str.replace(',', '.')
    df[column] = pd.to_numeric(df[column], errors= 'coerce')
```

```python
def add_labels(df):
  labels = []
  for precip in df['PRECIPITA√á√ÉO TOTAL, HOR√ÅRIO (mm)']:
    if precip < 4:
      labels.append('0 a 4 mm')
    elif precip < 8:
      labels.append('4 a 8 mm')
    elif precip < 12:
      labels.append('8 a 12 mm')
    elif precip < 16:
      labels.append('12 a 16 mm')
    elif precip < 20:
      labels.append('16 a 20 mm')
    elif precip < 24:
      labels.append('20 a 24 mm')
    else:
      labels.append('24 a 28 mm')
  df['labels'] = labels

add_labels(df)
```

```python
df.drop(['Data', 'Hora UTC','Unnamed: 19','RADIACAO GLOBAL (Kj/m¬≤)'],inplace=True, axis=1)
df.dropna(inplace=True)
```

1. Retire uma amostra Estratificada, use a label como refer√™ncia.

```python
# Retirar uma amostra estratificada com base na label
def stratified_sample(df, stratify_column, frac):
    return df.groupby(stratify_column, group_keys=False).apply(lambda x: x.sample(frac=frac))

# Exemplo: Retirando uma amostra estratificada de 10% de cada grupo
stratified_sample_df = stratified_sample(df, 'labels', 0.1)

# Exibir a amostra estratificada
stratified_sample_df.head()
```

6. Fa√ßa a normaliza√ß√£o e a padroniza√ß√£o de df["VENTO, DIRE√á√ÉO HORARIA (gr) (¬∞ (gr))"], e mostre os gr√°ficos da distribui√ß√£o antes e ap√≥s procedimento os procedimentos. Podemos dizer que √© uma distribui√ß√£o normal ? Justifique.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from scipy.stats import normaltest

# Coluna de interesse
wind_direction = df["VENTO, DIRE√á√ÉO HORARIA (gr) (¬∞ (gr))"].dropna()

# Normaliza√ß√£o
scaler_minmax = MinMaxScaler()
wind_direction_normalized = scaler_minmax.fit_transform(wind_direction.values.reshape(-1, 1))

# Padroniza√ß√£o
scaler_standard = StandardScaler()
wind_direction_standardized = scaler_standard.fit_transform(wind_direction.values.reshape(-1, 1))

# Plotando as distribui√ß√µes
plt.figure(figsize=(15, 5))

# Distribui√ß√£o original
plt.subplot(1, 3, 1)
sns.histplot(wind_direction, kde=True, color='blue', bins=30)
plt.title('Distribui√ß√£o Original')
plt.xlabel('Vento, Dire√ß√£o Hor√°ria (¬∞)')
plt.ylabel('Densidade')

# Distribui√ß√£o Normalizada
plt.subplot(1, 3, 2)
sns.histplot(wind_direction_normalized, kde=True, color='green', bins=30)
plt.title('Distribui√ß√£o Normalizada')
plt.xlabel('Vento, Dire√ß√£o Hor√°ria (Normalizado)')
plt.ylabel('Densidade')
plt.legend([],[], frameon=False)

# Distribui√ß√£o Padronizada
plt.subplot(1, 3, 3)
sns.histplot(wind_direction_standardized, kde=True, color='red', bins=30)
plt.title('Distribui√ß√£o Padronizada')
plt.xlabel('Vento, Dire√ß√£o Hor√°ria (Padronizado)')
plt.ylabel('Densidade')
plt.legend([],[], frameon=False)

plt.tight_layout()
plt.show()

# Verificando se a distribui√ß√£o √© normal
stat, p = normaltest(wind_direction_standardized)
stat = stat[0]  # Garantir que √© um escalar
p = p[0]        # Garantir que √© um escalar
print('Estat√≠stica do teste de normalidade: %.3f, p-valor: %.3f' % (stat, p))

if p > 0.05:
    print('A distribui√ß√£o parece ser normal (n√£o rejeitamos H0)')
else:
    print('A distribui√ß√£o n√£o parece ser normal (rejeitamos H0)')

# Estat√≠sticas Descritivas
print('\nDados Originais:')
print('M√©dia:', np.mean(wind_direction))
print('Desvio padr√£o:', np.std(wind_direction))
print('Vari√¢ncia:', np.var(wind_direction))

print('\nDados Padronizados:')
print('M√©dia:', np.mean(wind_direction_standardized))
print('Desvio padr√£o:', np.std(wind_direction_standardized))
print('Vari√¢ncia:', np.var(wind_direction_standardized))
```
