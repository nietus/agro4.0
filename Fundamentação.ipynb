```python
zip_file_name = "/content/2024.zip"
import pandas as pd
import zipfile

with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    with zip_ref.open('INMET_S_SC_A895_CHAPECO_01-01-2024_A_30-06-2024.CSV') as csv_file:

        df = pd.read_csv(csv_file, sep=';',skiprows=8, encoding='latin-1',on_bad_lines='skip')
```

```python
for column in df:
  if df[column].dtype == 'object':
    df[column] = df[column].str.replace(',', '.')
    df[column] = pd.to_numeric(df[column], errors= 'coerce')
```

```python
def add_labels(df):
  labels = []
  for precip in df['PRECIPITAÃ‡ÃƒO TOTAL, HORÃRIO (mm)']:
    if precip < 4:
      labels.append('0 a 4 mm')
    elif precip < 8:
      labels.append('4 a 8 mm')
    elif precip < 12:
      labels.append('8 a 12 mm')
    elif precip < 16:
      labels.append('12 a 16 mm')
    elif precip < 20:
      labels.append('16 a 20 mm')
    elif precip < 24:
      labels.append('20 a 24 mm')
    else:
      labels.append('24 a 28 mm')
  df['labels'] = labels

add_labels(df)
```

1. Retire uma amostra Estratificada, use a label como referÃªncia.

```python
# Retirar uma amostra estratificada com base na label
def stratified_sample(df, stratify_column, frac):
    return df.groupby(stratify_column, group_keys=False).apply(lambda x: x.sample(frac=frac))

# Exemplo: Retirando uma amostra estratificada de 10% de cada grupo
stratified_sample_df = stratified_sample(df, 'labels', 0.1)

# Exibir a amostra estratificada
stratified_sample_df.head()
```

6. FaÃ§a a normalizaÃ§Ã£o e a padronizaÃ§Ã£o de df["VENTO, DIREÃ‡ÃƒO HORARIA (gr) (Â° (gr))"], e mostre os grÃ¡ficos da distribuiÃ§Ã£o antes e apÃ³s procedimento os procedimentos. Podemos dizer que Ã© uma distribuiÃ§Ã£o normal ? Justifique.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from scipy.stats import normaltest

# Coluna de interesse
wind_direction = df["VENTO, DIREÃ‡ÃƒO HORARIA (gr) (Â° (gr))"].dropna()

# NormalizaÃ§Ã£o
scaler_minmax = MinMaxScaler()
wind_direction_normalized = scaler_minmax.fit_transform(wind_direction.values.reshape(-1, 1))

# PadronizaÃ§Ã£o
scaler_standard = StandardScaler()
wind_direction_standardized = scaler_standard.fit_transform(wind_direction.values.reshape(-1, 1))

# Plotando as distribuiÃ§Ãµes
plt.figure(figsize=(15, 5))

# DistribuiÃ§Ã£o original
plt.subplot(1, 3, 1)
sns.histplot(wind_direction, kde=True, color='blue', bins=30)
plt.title('DistribuiÃ§Ã£o Original')
plt.xlabel('Vento, DireÃ§Ã£o HorÃ¡ria (Â°)')
plt.ylabel('Densidade')

# DistribuiÃ§Ã£o Normalizada
plt.subplot(1, 3, 2)
sns.histplot(wind_direction_normalized, kde=True, color='green', bins=30)
plt.title('DistribuiÃ§Ã£o Normalizada')
plt.xlabel('Vento, DireÃ§Ã£o HorÃ¡ria (Normalizado)')
plt.ylabel('Densidade')
plt.legend([],[], frameon=False)

# DistribuiÃ§Ã£o Padronizada
plt.subplot(1, 3, 3)
sns.histplot(wind_direction_standardized, kde=True, color='red', bins=30)
plt.title('DistribuiÃ§Ã£o Padronizada')
plt.xlabel('Vento, DireÃ§Ã£o HorÃ¡ria (Padronizado)')
plt.ylabel('Densidade')
plt.legend([],[], frameon=False)

plt.tight_layout()
plt.show()

# Verificando se a distribuiÃ§Ã£o Ã© normal
stat, p = normaltest(wind_direction_standardized)
stat = stat[0]  # Garantir que Ã© um escalar
p = p[0]        # Garantir que Ã© um escalar
print('EstatÃ­stica do teste de normalidade: %.3f, p-valor: %.3f' % (stat, p))

if p > 0.05:
    print('A distribuiÃ§Ã£o parece ser normal (nÃ£o rejeitamos H0)')
else:
    print('A distribuiÃ§Ã£o nÃ£o parece ser normal (rejeitamos H0)')

# EstatÃ­sticas Descritivas
print('\nDados Originais:')
print('MÃ©dia:', np.mean(wind_direction))
print('Desvio padrÃ£o:', np.std(wind_direction))
print('VariÃ¢ncia:', np.var(wind_direction))

print('\nDados Padronizados:')
print('MÃ©dia:', np.mean(wind_direction_standardized))
print('Desvio padrÃ£o:', np.std(wind_direction_standardized))
print('VariÃ¢ncia:', np.var(wind_direction_standardized))
```
